```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
import spacy

from matplotlib.pyplot import imread
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline
```

```python
# load the dataset
df=pd.read_csv('Elon_musk.csv',encoding='Latin-1')
df
```

```python
df1=df.drop("Unnamed: 0",axis=1)
```

```python
df1
```

# Text preprocessing

```python
df1=[Text.strip() for Text in df1.Text] # remove both the leading and the trailing characters
df1=[Text for Text in df1 if Text] # removes empty strings, because they are considered in Python as False
df1[0:12]
```

```python
# Joining the list into one string/text
df1_text=' '.join(df1)
df1_text
```

```python
# remove Twitter username handles from a given twitter text. (Removes @usernames)
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer(strip_handles=True)
tweets_tokens=tknzr.tokenize(df1_text)
print(tweets_tokens)

# remove the username( @xyz ) from the tweets. ex @kunalb11 is removed from the text
```

```python
# Again Joining the list into one string/text
tweets_tokens_text=' '.join(tweets_tokens)
tweets_tokens_text
```

```python
# Remove Punctuations 
no_punc_text=tweets_tokens_text.translate(str.maketrans('','',string.punctuation))
no_punc_text
```

```python
# remove https or url within text
import re
no_url_text=re.sub(r'http\S+', '', no_punc_text)
no_url_text
```

```python
from nltk.tokenize import word_tokenize
text_tokens=word_tokenize(no_url_text)
print(text_tokens)
```

```python
# Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Tokens count
len(text_tokens)
```

```python
# Remove Stopwords
from nltk.corpus import stopwords
my_stop_words=stopwords.words('english')

newstop_word_list = ['\x92','rt','ye','yeah','haha','Yes','U0001F923','I']
my_stop_words.extend(newstop_word_list)

no_stop_tokens=[word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens)
```

```python
# we append/increased the stop words by inserting 
# newstop_word_list = ['\x92','rt','ye','yeah','haha','Yes','U0001F923','I']
```

```python
# Normalize the data
lower_words=[Text.lower() for Text in no_stop_tokens]
print(lower_words[100:200])
```

```python
# normalize the text means
# remove all punctuations mark
# transform all words to lower case
# replace synonymn with more common general word by using a dictionary such as Wordnet
```

```python
# Stemming (Optional)
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_tokens=[ps.stem(word) for word in lower_words]
print(stemmed_tokens[100:200])
```

```python
# stemming
# chops off the end of words without considering linguistic features
# once becomes onc 
# reasonably becomes reason
```

```python
# Lemmatization
nlp=spacy.load('en_core_web_sm')
doc=nlp(' '.join(lower_words))
print(doc)
```

# Lemmiatization
convert into base form or dictionary form of word 

```python
lemmas=[token.lemma_ for token in doc]
print(lemmas)
```

```python
clean_tweets=' '.join(lemmas)
clean_tweets
```

# Named entity recognizition

```python
# Parts Of Speech (POS) Tagging
nlp=spacy.load('en_core_web_sm')

one_block=clean_tweets
doc_block=nlp(one_block)
spacy.displacy.render(doc_block,style='ent',jupyter=True)
```

```python
for token in doc_block[150:250]:
    print(token,token.pos_)    
```

```python
# Filtering the nouns and verbs only
nouns_verbs=[token.text for token in doc_block if token.pos_ in ('NOUN','VERB')]
print(nouns_verbs[150:250])
```

```python
# Counting the noun & verb tokens
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

X=cv.fit_transform(nouns_verbs)
sum_words=X.sum(axis=0)

words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key=lambda x: x[1], reverse=True)

wd_df=pd.DataFrame(words_freq)
wd_df.columns=['word','count']
wd_df[0:10] # viewing top ten results
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
import spacy
from wordcloud import WordCloud, STOPWORDS
```

```python
affin=pd.read_csv('Afinn.csv',sep=',',encoding='Latin-1')
affin
```

```python
affinity_scores = affin.set_index('word')['value'].to_dict()
```

```python
affinity_scores
```

```python
#Custom function :score each word in a sentence in lemmatised form, 
#but calculate the score for the whole original sentence.
nlp = spacy.load('en_core_web_sm')
sentiment_lexicon = affinity_scores

def calculate_sentiment(text: str = None):
    sent_score = 0
    if text:
        sentence = nlp(text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_, 0)
    return sent_score
```

```python
# test that it works
#calculate_sentiment(text = 'The class is awesome. But the topic is very sluggish.')
calculate_sentiment(text = 'I love coding, but it is very difficult sometimes, though it is interesting too')
```

```python
sent_df['sentiment_value'] = sent_df['sentence'].apply(calculate_sentiment)
sent_df
```
