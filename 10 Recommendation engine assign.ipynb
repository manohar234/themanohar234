# user id and book rating are numeric while book.Title is object so we can not apply +(add) , first we convert it into string

```python
df['User_ID'] = df['User_ID'].apply(str)
```

```python
df['Book_Rating'] = df['Book_Rating'].apply(str)
```

```python
df.info()
```

```python
def get_important_features(data):
    important_features = []
    for i in range(data.shape[0]):
        a = (data["User_ID"][i]+' '+data["Book_Title"][i]+' '+data["Book_Rating"][i])
        important_features.append(a)
        
    return important_features
```

```python
df_clean = df[columns]
```

```python
df_clean["important_features"] = get_important_features(df_clean) #  above defined function get_important_features(data) with
# return function
df_clean
```

```python
#Convert text into matrix
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer


cm = CountVectorizer().fit_transform(df_clean["important_features"])
```

```python
# countvectorizer is used to make vector matrix of each  unique words (from a given text) with their frequency....
 # but in index form as the code of unique words
```

```python
# CountVectorizer makes new columns having new words in column "important_features
```

```python
cs = cosine_similarity(cm) # one book compare with all other book , very similar to corelation .corr() of numerical columns
```
