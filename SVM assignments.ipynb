```python
df1= pd.get_dummies(df)
df1
```

```python
X = df.drop(['Salary'],axis=1)
```

# feature and target

```python
X = df1.drop(['Salary_ <=50K','Salary_ >50K'],axis=1)
y = df1["Salary_ >50K"]
```

```python
X
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

# train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

# Model creation

```python
from sklearn.svm import SVC
model = SVC(kernel="rbf",C=1.0,gamma="scale")
model.fit(X_train,y_train)
```

# Model validation

```python
y_pred = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

# kfold testing

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
estimator = SVC()
result = cross_val_score(estimator,X,y,cv=KFold(n_splits=5))
```

```python
result
```

# Grid search hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV,KFold,cross_val_score
```

```python
test_model = SVC()
param = {"kernel":["rbf","poly","sigmoid"]}

grid = GridSearchCV(test_model,cv=KFold(n_splits=2),param_grid=param)

grid_result = grid.fit(X,y)
```

```python
grid_result.best_params_
```

```python
grid_result.best_score_
```

```python
# it takes too much time to for performing hyperparameter tuning because 
# we have very high number of columns, 102 columns .
```

# Final model

```python
from sklearn.svm import SVC
model = SVC(kernel="rbf",C=1.0,gamma="scale")
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
