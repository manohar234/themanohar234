```python
# feature and target selection
```

```python
X = df.drop(['size_category','month','day'],axis=1)
y = df["size_category"]
```

```python
# as month and day are already given in binary format as monthdec etc,so we 
# removed month and day columns
```

```python
X.head()
```

```python
y.head()
```

```python
# standarization
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# TRain test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.svm import SVC
model = SVC(kernel="rbf",C=1.0,gamma="scale")
model.fit(X_train,y_train)
```

```python
# model validation
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# kfold testing
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
estimator = SVC()
result = cross_val_score(estimator,X,y,cv=KFold(n_splits=10))
result
```

```python
# gridsearch hyperparameter tuning
```

```python
from sklearn.model_selection import GridSearchCV,KFold,cross_val_score
```

```python
test_model = SVC()
param = {"kernel":["rbf","poly","sigmoid"],"C":[1,100,1000]}

grid = GridSearchCV(test_model,cv=KFold(n_splits=10),param_grid=param)

grid_result = grid.fit(X,y)
```

```python
grid_result.best_params_
```

```python
grid_result.best_score_
```

```python
# Final model
```

```python
from sklearn.svm import SVC
model = SVC(kernel="rbf",C=1000,gamma="scale")
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
