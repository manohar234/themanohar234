# Split features and target

```python
X=dff.iloc[:,0:12]
y=dff["ysales"].values
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler # standarization is not required for decision tree
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(X)
```

# Train and test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(x,y,train_size=0.75,random_state=0)
```

# Model Creation

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier()
model.fit(X_train,y_train)
```

# Model Validation: Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
cm
```

```python
print(classification_report(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
y_pred1 = model.predict(X_train)
accuracy_score(y_train,y_pred1)
```

# Model Visualization & Tree  Visualization

```python
from sklearn import tree
plt.figure(figsize=(25,15))
tree.plot_tree(model,filled=True)

plt.show()
```

# Pruning tree

```python
model2 = DecisionTreeClassifier(criterion="gini",max_depth=None,ccp_alpha=0.0)
```

```python
path = model2.cost_complexity_pruning_path(X_train,y_train)
```

```python
ccp_alphas = path.ccp_alphas 
```

```python
ccp_alphas
```

```python
len(ccp_alphas)
```

```python
train_score = []
test_score = []

for i in ccp_alphas:
    model1 = DecisionTreeClassifier(ccp_alpha=i)
    model1.fit(X_train,y_train)
    train_score.append(model1.score(X_train,y_train))
    test_score.append(model1.score(X_test,y_test))    
train_score
```

```python
model1.score(X_test,y_test) # model.score() gives accuracy of dataset similar to accuraccy_score()
```

```python
plt.scatter(ccp_alphas,train_score,color="green",label="Training")
plt.plot(ccp_alphas,train_score,color="green",label="Training")
plt.scatter(ccp_alphas,test_score,color = "red",label="Test")
plt.plot(ccp_alphas,test_score,color = "red",label="Test")
plt.xlabel("CCP alpha")
plt.ylabel("Accuracy")

plt.legend()
plt.show()
```

```python
plt.scatter(ccp_alphas,train_score,color="green",label="Training")
plt.plot(ccp_alphas,train_score,color="green",label="Training")
plt.scatter(ccp_alphas,test_score,color = "red",label="Test")
plt.plot(ccp_alphas,test_score,color = "red",label="Test")
plt.xlabel("CCP alpha")
plt.ylabel("Accuracy")
plt.xlim(0.02,0.04)
plt.legend()
plt.grid()
plt.show()
```
