```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("delivery_time.csv")
df
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.scatter(df["Sorting Time"],df["Delivery Time"])
plt.title("Relationship between sorting time & Delivery Time")
plt.xlabel("sorting type")
plt.ylabel("Delivery Time")
plt.show()
```

```python
X = df[["Sorting Time"]]
Y = df["Delivery Time"]
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(X)
y_pred
```

```python
pd.DataFrame({"Actual":Y,"Predicted":y_pred,"Residual":Y-y_pred})
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y,y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(Y,y_pred)
```

```python
model.predict([[7]])
```

```python
plt.scatter(df["Sorting Time"],df["Delivery Time"],color='green',label="Actual")
plt.scatter(df["Sorting Time"],y_pred,color='red',label="predicted")
plt.plot(df["Sorting Time"],y_pred,color='red',label="predicted")

plt.title("Relationship between Sorting Time and delivery time")
plt.xlabel("Sorting Time")
plt.legend()
plt.ylabel("delivery time")
plt.show()
```

```python
df=pd.read_csv("Salary_Data.csv")
df
```

```python
plt.scatter(df["YearsExperience"],df["Salary"])
plt.title("Relationship between Age & Salary")
plt.xlabel("Exp")
plt.ylabel("Salary")
plt.show()
```

```python
X = df[["YearsExperience"]] # X should be in dataframe to use linear regression model of sciket learn
Y = df["Salary"]
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(X)
y_pred
```

```python
pd.DataFrame({"Actual":Y,"Predicted":y_pred,"Residual":Y-y_pred})
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y,y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(Y,y_pred)
```

```python
model.predict([[2.7]])
```

```python
plt.scatter(df["YearsExperience"],df["Salary"],color='green',label="Actual")
plt.scatter(df["YearsExperience"],y_pred,color='red',label="predicted")
plt.plot(df["YearsExperience"],y_pred,color='red',label="predicted")

plt.title("Relationship between Age & Salary")
plt.xlabel("Exp")
plt.legend()
plt.ylabel("Salary")
plt.show()
```
