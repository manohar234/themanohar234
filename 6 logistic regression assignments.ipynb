# data transformation

```python
df.head()
```

```python
# One-Hot Encoding of categrical variables
df1=pd.get_dummies(df,columns=['job','marital','education','contact','poutcome','month'])
df1
```

```python
df1.info()
```

```python
df1["default"] = df1["default"].replace({"yes":1,"no":0}) #labeling: LabelEncoder
df1["housing"] = df1["housing"].replace({"yes":1,"no":0})
df1["loan"] = df1["loan"].replace({"yes":1,"no":0})
df1["y"] = df1["y"].replace({"yes":1,"no":0})
df1
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
df1
```

# Feature and Target selection

```python
import pandas as pd
```

```python
# Dividing our data into input and output variables
x=pd.concat([df1.iloc[:,0:10],df1.iloc[:,11:]],axis=1)
y=df1.iloc[:,10]
```

```python
Y=y.values # converting y into array for using boolean indexing of iloc
Y
```

# Standarization not required

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(x) # x is converted from dataframe to array and and 
# StandardScaler assigns the z score of each value of
# dataframe with transform(). transform only change the X not the orignal data frame x.
```

```python
# standardscaler is actually standrization of daata set that means z score, 
# standarization means to bring down all the features to a common scale without distorting
#the difference in the range of the values
```

```python
X
```

# TRAIN and test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,Y,train_size=0.75,random_state=0)
```

```python
# randomly distribute the training data and test data in 75:25 ,ingeneral it is taken 3:2 ...
# because we want new data apart from model building data( training data) to test the model .
# so as to find actual possible accuracy , if they are not seprated then model got the same data 
# for testing . then model shows higher accuracy than orignal condition so it becomes overfitted  data overfitted
# (because we don noot have other data to verify the model for same data set(used for training and testing)
# accurraacy will be nealy 100%)
# but in real user testing accuraccy will fall which is not good
```

# Model Creation

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```
