```python
car1=cars.rename(columns={"R&D Spend":"rdspend","Marketing Spend":"mspend"})
# model will not accept column name as R&D Spend  because there is space between them , we need to remove the space between 
# the column name of "Marketing spend " to mspend( no space is allowed) 
```

```python
car1.head()
```

```python
car1
```

```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('Profit~rdspend+Administration+mspend+State',data=car1).fit()
```

```python
model.params
```

```python
model.summary()
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

```python
car1.head()
```

# calculating vif

```python
rsq_rdspend = smf.ols('rdspend~Administration+mspend+State',data=car1).fit().rsquared  
vif_rdspend = 1/(1-rsq_rdspend) # 16.33

rsq_adminis = smf.ols('Administration~rdspend+mspend+State',data=car1).fit().rsquared  
vif_adminis = 1/(1-rsq_adminis) # 564.98

rsq_mspend = smf.ols('mspend~rdspend+Administration+State',data=car1).fit().rsquared  
vif_mspend = 1/(1-rsq_mspend) #  564.84


# Storing vif values in a data frame
d1 = {'Variables':['rdspend','Administration','mspend'],'VIF':[vif_rdspend,vif_adminis,vif_mspend]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line,line=q means line is fit through the quarantiles
# as quarantiles are present in boxplot (meaning 25 percent of data)
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>5000))
```

# cooks distance

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(car1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```
